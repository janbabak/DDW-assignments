## Apriory algorithm from the lecture

```python
from collections import Counter

def frequentItems(transactions, support):
    counter = Counter()
    for trans in transactions:
        counter.update(frozenset([t]) for t in trans)
    return set(item for item in counter if counter[item]/len(transactions) >= support), counter

def generateCandidates(L, k):
    candidates = set()
    for a in L:
        for b in L:
            union = a | b
            if len(union) == k and a != b:
                candidates.add(union)
    return candidates

def filterCandidates(transactions, itemsets, support):
    counter = Counter()
    for trans in transactions:
        subsets = [itemset for itemset in itemsets if itemset.issubset(trans)]
        counter.update(subsets)
    return set(item for item in counter if counter[item]/len(transactions) >= support), counter

def apriori(transactions, support):
    result = list()
    resultc = Counter()
    candidates, counter = frequentItems(transactions, support)
    result += candidates
    resultc += counter
    k = 2
    while candidates:
        candidates = generateCandidates(candidates, k)
        candidates,counter = filterCandidates(transactions, candidates, support)
        result += candidates
        resultc += counter
        k += 1
    resultc = {item:(resultc[item]/len(transactions)) for item in resultc}
    return result, resultc
```

```python

frequentItemsets, supports = apriori(dataset, 0.1)
print("Result of Apriori algorithm:")
for f in frequentItemsets:
    print(f"{f}, {supports[f]}")
```

## Generate rules from apriori algorithm

- Here is a list of rules, which meet the support and confidence criteria.
- I set the confidence greater or equal to 0.5 and the support to 0.1.

```python
def generateRules(frequentItemsets, supports, minConfidence):
    result = []
    for fset in frequentItemsets:
        for consequent in fset:
            antecedent = set(fset).difference([consequent])
            support = supports[fset]
            confidence = supports[fset]
            if len(antecedent) > 0:
                confidence = support/supports[frozenset(antecedent)]
            if confidence > minConfidence:
                result.append({
                    "entecedent": antecedent,
                    "consequent": consequent,
                    "support": support,
                    "confidence": confidence,
                    "len": len(fset)
                })
                print(f"{antecedent} => {consequent}, - support: {round(support, 3)}, confidence: {round(confidence, 3)}")
    return pd.DataFrame(result)

rules = generateRules(frequentItemsets, supports, 0.5)
```
